## 1. Test of independence between product category and different parameters

### 1.1 Test of independence between product category and gender

```python
import pandas as pd
from scipy import stats
import numpy as np

df=pd.read_csv('blackfriday_clean.csv',index_col=0)
df.head()
```

```python
pc_by_gender=pd.melt(df,id_vars='gender',value_vars=['product_category_1','product_category_2','product_category_3'])
pc_by_gender.head()
```

```python
pc_by_gender_count=pc_by_gender.pivot_table(index='value',columns='gender',aggfunc='count')
pc_by_gender_count.head()
```

```python
chi2, p, dof, ex=stats.chi2_contingency(pc_by_gender_count)
print("The observed chi-square value is", chi2, ", the p-value is",p)
```

The expected frequencies in all cells in the contingency table are larger than 5, therefore the cases meet the sample size condition for Chi-squared test.

As computed, p-value is 0, the null hypothesis is rejected, so product category is dependent on the gender.

## 2. Test of independence between purchase level and different parameters

### 2.1 Categorize purchase levels for statistical analysis

```python
df_pa=df.pivot_table(values='purchase',index=['user_id','gender','age','occupation','city_category','stay_in_current_city_years','marital_status'],aggfunc=np.sum)
df_pa=df_pa.reset_index()
df_pa.head()
```

```python
max(df_pa['purchase'])
```

```python
min(df_pa['purchase'])
```

In order to conduct the Chi-square test of independence between purchase levels and different parameters, total purchase amounts are categorized into different categories. In this case, categories are chosen based on every 10000 interval of purchase amounts, 0, 1, 2, 3, 4 correspond to purchase amounts of 0-1,000,000, 1,000,000-2,000,000, 2,000,000-3,000,000, 3,000,000-4,000,000, above 4,000,000, repectively.

```python
df_pa['purchase level']=[v//1000000 if v<4000000 else 4 for v in df_pa['purchase']]
print(df_pa['purchase level'].value_counts())
```

### 2.2 Test of independence between purchase level and gender  

```python
pa_by_gender=pd.melt(df_pa,id_vars='gender',value_vars='purchase level')
pa_by_gender.head()
```

```python
pa_by_gender_count=pc_by_gender.pivot_table(index='value',columns='gender',aggfunc='count')
pa_by_gender_count.head()
```

```python
chi2, p, dof, ex=stats.chi2_contingency(pa_by_gender_count)
print("The observed chi-square value is", chi2, ", the p-value is",p)
```

The expected frequencies in all cells in the contingency table are larger than 5, therefore the cases meet the sample size condition for Chi-squared test.

As computed, p-value is very small, the null hypothesis is rejected, so purchase level is dependent on gender.
